### percentage of missing values in per column

```python
df.isna().sum().sort_values()# Ascending
```

```python
df.isna().sum().sort_values(ascending=False)# descending
```

```python
missing_percentages=df.isna().sum().sort_values(ascending=False)/len(df)
missing_percentages
```

### Visual representation of missing values 

```python
### removing The non zero columns
only_missing_columns=missing_percentages[missing_percentages !=0]
```

```python
only_missing_columns
```

```python
only_missing_columns.plot(kind='barh')
```

```python
# To insert the mean value of each column into its missing rows:
df.fillna(df.mean(numeric_only=True).round(1), inplace=True)

# For median:
df.fillna(df.median(numeric_only=True).round(1), inplace=True)

print(df)
```
